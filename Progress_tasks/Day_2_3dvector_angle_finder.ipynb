### Day 2 task: Program that takes in 2 randomised 3d vectors and computes the angle between the 2 vectors
**Formula used = cosθ = a.b / |a||b|**
Turns into θ = cos-1(a.b / |a||b|)

|a| and |b| are the magnitude of / absolute value
    |a| = (ax, ay, az) -> These are values from the x, y, z axis of the vector
**How to calculate:**
    |a| = √ax² + ay² + az²

**How to calculate dot product:**
    eg vectors
    a = [0, 0, 8]
    b = [3, 8, 6]
    a.b = (0 x 3) + (0 x 8) + (8 x 6) = 48

```python
import numpy as np

def calculate_vector_angle(v1, v2):
    v1 = np.array(v1)
    v2 = np.array(v2)
    dot_product = np.dot(v1, v2)
    abs_v1 = np.linalg.norm(v1)
    abs_v2 = np.linalg.norm(v2)
    angle_radians = np.arccos(dot_product / (abs_v1 * abs_v2))
    angle_degrees = np.degrees(angle_radians)
    print(f"Angle between vectors: {angle_degrees:.5f} degrees")

def get_random_vector(dimensions):
    print( np.random.randint(0, 10, size=dimensions))
    return np.random.randint(0, 10, size=dimensions)

calculate_vector_angle(get_random_vector(3), get_random_vector(3))
```
